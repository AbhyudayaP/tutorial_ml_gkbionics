```python
from sklearn import datasets
```

```python
X, y = datasets.make_classification(n_features=2, n_redundant=0, random_state=5)
plt.figsize(5, 5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.prism)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xticks(())
plt.yticks(())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X, y)
```

```python
plt.figsize(5, 5)
w = logreg.coef_.ravel()
a = -w[0] / w[1]
xx = np.linspace(-3, 3)
yy = a * xx - logreg.intercept_ / w[1]
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.prism)
plt.plot(xx, yy)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xticks(())
plt.yticks(())
plt.savefig("presentation/logreg-pics/synthetic_line.pdf")
```
