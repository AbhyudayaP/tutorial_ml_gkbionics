```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(noise=0.2, n_samples=200)
```

```python
X.shape
```

```python
y.shape
```

```python
print(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=100)
plt.figsize(14,10)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
print(knn)
```

```python
knn.fit(X_train, y_train)
```

```python
y_predict = knn.predict(X_test)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=100)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_predict, alpha=0.3, s=100)
#plt.scatter(X_test[:, 0], X_test[:, 1], c='w', alpha=0.3, s=100)
plt.figsize(14,10)
```

```python

y_predict
```

```python
y_test
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()
```

```python
svm.fit(X_train, y_train)
```

```python
y_predict = svm.predict(X_test)
```

```python
plt.figsize(14,10)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=100)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_predict, alpha=0.3, s=100)
w = svm.coef_.ravel()
a = -w[0] / w[1]
xx = np.linspace(-1, 2)
yy = a * xx - svm.intercept_ / w[1]
plt.plot(xx, yy)

```

```python
import pandas as pd
train_data = pd.read_csv("../kaggle_insults/train.csv")
test_data = pd.read_csv("../kaggle_insults/test_with_solutions.csv")
```
